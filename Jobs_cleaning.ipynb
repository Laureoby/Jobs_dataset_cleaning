```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
df = pd.read_csv('Uncleaned_DS_jobs.csv')
df
```

```python
#we search if dataset has Nan values
df.isnull().sum()
```

We will work on some column like salary estimate and others

```python
#function which extract numbers and hyphen in Salary Estimate column
def extract_numbers(text):
    return re.sub(r'[^\d\-]',"", str(text))
df['Salary Estimate'] = df['Salary Estimate'].apply(extract_numbers)
```

```python
df
```

```python
#We divide salary estimate in two column min_salary and max_salary
def min_max(salary):
  match = re.search(r'(\d+)-(\d+)', salary)
  if match:
    return int(match.group(1)), int(match.group(2))
  else:
    return None

df[['min_salary', 'max_salary']] = df['Salary Estimate'].apply(lambda x: pd.Series(min_max(x)))
df.head()
```

```python
def extract_after_newline(text):
    if '\n' in str(text):
        return str(text).split('\n')[0]
    else:
        return text
df['Company Name'] = df['Company Name'].apply(extract_after_newline)
df.head()
```

```python
df['Same location'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis=1)
df.head()
```

```python
df['Company age'] = 2024 - df['Founded']
df.head()
```

```python
#creating column job_simp
def job_simplifier(title):
  if 'data scientist' in title.lower():
    return 'data_scientist'
  elif 'analyst' in title.lower():
    return 'analyst'
  elif 'machine learning engineer' in title.lower():
    return 'mle'
  else:
    return 'na'

df['job_simp'] = df['Job Title'].apply(job_simplifier)
df.head(10)
```

```python
cols = ['index', 'Competitors']
df.drop(cols, axis=1, inplace=True)
```

```python
df.drop('Salary Estimate', axis=1, inplace=True)
```

```python
df['average'] = df.apply(lambda x: (x['min_salary'] + x['max_salary'])/2, axis=1)
df
```

```python
#creating column for specialities
def fill_column(df, search_value, column_name):
    df[column_name] = df['Job Description'].apply(lambda x:1 if search_value.lower() in str(x).lower() else 0)
    return df


df = fill_column(df, 'python', 'python')
df = fill_column(df, 'excel', 'excel')
df = fill_column(df, 'hadoop', 'hadoop')
df = fill_column(df, 'spark', 'spark')
df = fill_column(df, 'aws', 'aws')
df = fill_column(df, 'tableau', 'tableau')
df = fill_column(df, 'big_data', 'big_data')

#print the dataframe
df
```

```python
# Calcul de la moyenne de la colonne 'rating', en excluant les valeurs -1
mean_value = df[df['Rating'] != -1]['Rating'].mean()
# Remplacement des valeurs -1 par la moyenne dans la colonne 'rating'
df['Rating'].replace(-1, mean_value, inplace=True)
```

```python
# function which extract minimum and maximum values
def extract_min_max(Size):
    if 'to' in Size:
        min_val, max_val = Size.split(' to ')
        min_val = int(min_val)
        max_val = int(max_val.split(' ')[0])
    elif '+' in Size:
        min_val = int(Size.split('+')[0])
        max_val = float('inf')  # Use float('inf') to represent "more than"
    else:
        min_val = max_val = None  # for 'Unknown' or others non numeric values
    return pd.Series([min_val, max_val])

# Apply function and create new columns
df[['Min_Employees', 'Max_Employees']] = df['Size'].apply(extract_min_max)

# print updated dataframe
df
```

```python
df.drop('Size', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
#we replace Nan values in Min and Max employees by mean
# Calcul de la moyenne de la colonne 'Max_Employees' et 'Min_Employees', en excluant les NaN
mean_min=df['Min_Employees'].mean()
mean_max=df['Max_Employees'].mean()
# Remplacement des valeurs NaN par la moyenne dans la colonne 'Min_Employees' et 'Max_Employees'
df['Min_Employees'].fillna(value=mean_min, inplace=True)
df['Max_Employees'].fillna(value=mean_max, inplace=True)
df.isnull().sum()
```

```python
#We prepare dataset
from sklearn.preprocessing import LabelEncoder
# Initialisation des encodeurs et des dictionnaires pour chaque colonne
columns_to_encode = ['Job Title', 'Company Name', 'Location', 'Headquarters', 'Type of ownership', 'Industry', 'Sector', 'job_simp']
encoders = {col: LabelEncoder() for col in columns_to_encode}
label_mapping = {}

# Application de LabelEncoder à chaque colonne et stockage du mappage
for col in columns_to_encode:
    df[col] = encoders[col].fit_transform(df[col])
    label_mapping[col] = {index: label for index, label in enumerate(encoders[col].classes_)}
```
